```python
###SVM
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei']
import scipy as sp
%matplotlib inline
%config inlinebackend.figure_format="retina"
from matplotlib.font_manager import FontProperties
from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm
import matplotlib as mpl
import seaborn as sns
from sklearn.svm import SVC,LinearSVC
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
from sklearn.datasets import make_classification
import graphviz
import pydotplus
from io import StringIO
from IPython.display import Image

data=pd.read_csv("*:/****/Ternary_Glassformation_dataset.csv")
train,test = train_test_split(data, test_size=0.2)
print(data.shape)
print(train.shape)
print(test.shape)
data_pro=[train,test]
print("data.info：\n",data.info())
print(data.columns.values)
Y1=['GFA']
X1=['a1','a2','a3','Ri1','Ri2','Ri3','E1','E2','E3','Tb1','Tb2','Tb3','Tm1','Tm2','Tm3','ρ1','ρ2','ρ3','Aw1','Aw2','Aw3', 'Rc1','Rc2','Rc3','Ra1','Ra2','Ra3','IE1','IE2','IE3','Eea1','Eea2','Eea3','Ia1','Ia2','Ib1','Ib2','Ic1','Ic2','α1','α2','α3','Cp1','Cp2','Cp3','λ1','λ2','λ3','σ1','σ2','σ3','Hf1','Hf2','Hf3','Hv1','Hv2','Hv3',
'χ1','χ2','χ3','AN1','AN2','AN3','P1','P2','P3','GP1','GP2','GP3','VEC1','VEC2','VEC3','sVEC1','sVEC2','sVEC3','pVEC1','pVEC2','pVEC3','dVEC1','dVEC2','dVEC3','W1','W2','W3','LP1','LP2','LP3','∆H12','∆H13','∆H23']
print(train[Y1].shape)
print(train[X1].shape)
Y2=['GFA']
X2=['Ri','E','Tb','Tm','ρ','Aw', 'Rc','Ra','IE','Eea','I1','I2','α','Cp','λ','σ','Hf','Hv','χ','AN','P','GP','VEC','sVEC','pVEC','dVEC','W','LP','∆Hmix','∆Sc','δ','VA']
print(train[Y2].shape)
print(train[X2].shape)
Y3=['GFA']
X3=['Ri','E','Tb','Aw','Rc','IE','Eea','α','σ','Hf','GP','sVEC','pVEC','dVEC','LP','∆Hmix','∆Sc','δ']
print(train[Y3].shape)
print(train[X3].shape)
Y4=['GFA']
X4=['E','ρ','Aw','Eea','α','Cp','λ','σ','Hf','AN','VEC','pVEC','dVEC','LP','∆Hmix','∆Sc','δ','VA']
print(train[Y4].shape)
print(train[X4].shape)
Y5=['GFA']
X5=['Aw','Eea','α','σ','Hf','pVEC','dVEC','LP','∆Hmix','∆Sc','δ','VA',]
print(train[Y5].shape)
print(train[X5].shape)
traindata_x1,valdata_x1,traindata_y1,valdata_y1=train_test_split(train[X1],train[Y1],test_size=0.25)
print(traindata_x1.shape)
print(valdata_x1.shape)
traindata_x2,valdata_x2,traindata_y2,valdata_y2=train_test_split(train[X2],train[Y2],test_size=0.25)
print(traindata_x2.shape)
print(valdata_x2.shape)
traindata_x3,valdata_x3,traindata_y3,valdata_y3=train_test_split(train[X3],train[Y3],test_size=0.25)
print(traindata_x3.shape)
print(valdata_x3.shape)
traindata_x4,valdata_x4,traindata_y4,valdata_y4=train_test_split(train[X4],train[Y4],test_size=0.25)
print(traindata_x4.shape)
print(valdata_x4.shape)
traindata_x5,valdata_x5,traindata_y5,valdata_y5=train_test_split(train[X5],train[Y5],test_size=0.25)
print(traindata_x5.shape)
print(valdata_x5.shape)

#SVM ALGORITHM
rbfsvm1=SVC(probability=True)
rbfsvm1.fit(traindata_x1,traindata_y1)
rbfsvm2=SVC(probability=True)
rbfsvm2.fit(traindata_x2,traindata_y2)
rbfsvm3=SVC(probability=True)
rbfsvm3.fit(traindata_x3,traindata_y3)
rbfsvm4=SVC(probability=True)
rbfsvm4.fit(traindata_x4,traindata_y4)
rbfsvm5=SVC(probability=True)
rbfsvm5.fit(traindata_x5,traindata_y5)
from sklearn import metrics 
valdatapre_y1_1=rbfsvm1.predict(valdata_x1)
print(metrics.accuracy_score(valdata_y1,valdatapre_y1_1))
print(metrics.classification_report(valdata_y1,valdatapre_y1_1))
valdatapre_y2_1=rbfsvm2.predict(valdata_x2)
print(metrics.accuracy_score(valdata_y2,valdatapre_y2_1))
print(metrics.classification_report(valdata_y2,valdatapre_y2_1))
valdatapre_y3_1=rbfsvm3.predict(valdata_x3)
print(metrics.accuracy_score(valdata_y3,valdatapre_y3_1))
print(metrics.classification_report(valdata_y3,valdatapre_y3_1))
valdatapre_y4_1=rbfsvm4.predict(valdata_x4)
print(metrics.accuracy_score(valdata_y4,valdatapre_y4_1))
print(metrics.classification_report(valdata_y4,valdatapre_y4_1))
valdatapre_y5_1=rbfsvm5.predict(valdata_x5)
print(metrics.accuracy_score(valdata_y5,valdatapre_y5_1))
print(metrics.classification_report(valdata_y5,valdatapre_y5_1))
#PLOT ROC CURVE
valdatapre_y1_1=rbfsvm1.predict(valdata_x1)
valdatay__proba1=rbfsvm1.predict_proba(valdata_x1)[:,1]
fpr_rbfsvm1,tpr_rbfsvm1,thr =metrics.roc_curve(valdata_y1,valdatay__proba1, pos_label=1)
auc=metrics.auc(fpr_rbfsvm1,tpr_rbfsvm1)
plt.figure(figsize=(6,6))
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr_rbfsvm1,tpr_rbfsvm1,"r",linewidth=1.5)
plt.xlabel("False Positive Rate",fontsize=15)
plt.ylabel("True Positive Rate",fontsize=15)
plt.xlim(0,1)
plt.ylim(0,1)
plt.title("rbfsvm1 ROC curve",fontsize=15)
plt.text(0.2,0.8,"auc="+str(round(auc,4)),fontsize=15)
plt.show()
valdatapre_y2_1=rbfsvm2.predict(valdata_x2)
valdatay__proba2=rbfsvm2.predict_proba(valdata_x2)[:,1]
fpr_rbfsvm2,tpr_rbfsvm2,thr =metrics.roc_curve(valdata_y2,valdatay__proba2, pos_label=1)
auc=metrics.auc(fpr_rbfsvm2,tpr_rbfsvm2)
plt.figure(figsize=(6,6))
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr_rbfsvm2,tpr_rbfsvm2,"r",linewidth=1.5)
plt.xlabel("False Positive Rate",fontsize=15)
plt.ylabel("True Positive Rate",fontsize=15)
plt.xlim(0,1)
plt.ylim(0,1)
plt.title("rbfsvm2 ROC curve",fontsize=15)
plt.text(0.2,0.8,"auc="+str(round(auc,4)),fontsize=15)
plt.show()
valdatapre_y3_1=rbfsvm3.predict(valdata_x3)
valdatay__proba3=rbfsvm3.predict_proba(valdata_x3)[:,1]
fpr_rbfsvm3,tpr_rbfsvm3,thr =metrics.roc_curve(valdata_y3,valdatay__proba3, pos_label=1)
auc=metrics.auc(fpr_rbfsvm3,tpr_rbfsvm3)
plt.figure(figsize=(6,6))
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr_rbfsvm3,tpr_rbfsvm3,"r",linewidth=1.5)
plt.xlabel("False Positive Rate",fontsize=15)
plt.ylabel("True Positive Rate",fontsize=15)
plt.xlim(0,1)
plt.ylim(0,1)
plt.title("rbfsvm3 ROC curve",fontsize=15)
plt.text(0.2,0.8,"auc="+str(round(auc,4)),fontsize=15)
plt.show()
valdatapre_y4_1=rbfsvm4.predict(valdata_x4)
valdatay__proba4=rbfsvm4.predict_proba(valdata_x4)[:,1]
fpr_rbfsvm4,tpr_rbfsvm4,thr =metrics.roc_curve(valdata_y4,valdatay__proba4, pos_label=1)
auc=metrics.auc(fpr_rbfsvm4,tpr_rbfsvm4)
plt.figure(figsize=(6,6))
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr_rbfsvm4,tpr_rbfsvm4,"r",linewidth=1.5)
plt.xlabel("False Positive Rate",fontsize=15)
plt.ylabel("True Positive Rate",fontsize=15)
plt.xlim(0,1)
plt.ylim(0,1)
plt.title("rbfsvm4 ROC curve",fontsize=15)
plt.text(0.2,0.8,"auc="+str(round(auc,4)),fontsize=15)
plt.show()
valdatapre_y5_1=rbfsvm5.predict(valdata_x5)
valdatay__proba5=rbfsvm5.predict_proba(valdata_x5)[:,1]
fpr_rbfsvm5,tpr_rbfsvm5,thr =metrics.roc_curve(valdata_y5,valdatay__proba5, pos_label=1)
auc=metrics.auc(fpr_rbfsvm5,tpr_rbfsvm5)
plt.figure(figsize=(6,6))
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr_rbfsvm5,tpr_rbfsvm5,"r",linewidth=1.5)
plt.xlabel("False Positive Rate",fontsize=15)
plt.ylabel("True Positive Rate",fontsize=15)
plt.xlim(0,1)
plt.ylim(0,1)
plt.title("rbfsvm5 ROC curve",fontsize=15)
plt.text(0.2,0.8,"auc="+str(round(auc,4)),fontsize=15)
plt.show()

#GridSearchCV
from sklearn.model_selection import GridSearchCV
rbfSVM=SVC()
parameters={'kernel': ['rbf'], 'gamma': [0.00001,0.0001,0.001,0.01,0.1,1, 10, 100, 1000,10000,100000,1000000],'C': [0.00001,0.0001,0.001,0.01,0.1,1, 10, 100, 1000,10000,100000,1000000]}
rbfSVM1_GridSearch=GridSearchCV(rbfSVM,parameters,cv=5)
rbfSVM2_GridSearch=GridSearchCV(rbfSVM,parameters,cv=5)
rbfSVM3_GridSearch=GridSearchCV(rbfSVM,parameters,cv=5)
rbfSVM4_GridSearch=GridSearchCV(rbfSVM,parameters,cv=5)
rbfSVM5_GridSearch=GridSearchCV(rbfSVM,parameters,cv=5)
print(rbfSVM1_GridSearch.fit(traindata_x1,traindata_y1))
print(rbfSVM2_GridSearch.fit(traindata_x2,traindata_y2))
print(rbfSVM3_GridSearch.fit(traindata_x3,traindata_y3))
print(rbfSVM4_GridSearch.fit(traindata_x4,traindata_y4))
print(rbfSVM5_GridSearch.fit(traindata_x5,traindata_y5))
print("best_params：",rbfSVM1_GridSearch.best_params_)
print("best_score：",rbfSVM1_GridSearch.best_score_)
print("best_estimator：",rbfSVM1_GridSearch.best_estimator_)
print("best_params：",rbfSVM2_GridSearch.best_params_)
print("best_score：",rbfSVM2_GridSearch.best_score_)
print("best_estimator：",rbfSVM2_GridSearch.best_estimator_)
print("best_params：",rbfSVM3_GridSearch.best_params_)
print("best_score：",rbfSVM3_GridSearch.best_score_)
print("best_estimator：",rbfSVM3_GridSearch.best_estimator_)
print("best_params：",rbfSVM4_GridSearch.best_params_)
print("best_score：",rbfSVM4_GridSearch.best_score_)
print("best_estimator：",rbfSVM4_GridSearch.best_estimator_)
print("best_params：",rbfSVM5_GridSearch.best_params_)
print("best_score：",rbfSVM5_GridSearch.best_score_)
print("best_estimator：",rbfSVM5_GridSearch.best_estimator_)
#RETRAING WITH TUNED PARAMETERS
rbfSVM1_1=SVC(probability=True,C=1000,gamma=100,kernel="rbf",class_weight='balanced')
rbfSVM1_1=rbfSVM1_1.fit(train[X1],train[Y1])
rbfSVM2_1=SVC(probability=True,C=10,gamma=0.0001,kernel="rbf",class_weight='balanced')
rbfSVM2_1=rbfSVM2_1.fit(train[X2],train[Y2])
rbfSVM3_1=SVC(probability=True,C=10000,gamma=1e-05,kernel="rbf",class_weight='balanced')
rbfSVM3_1=rbfSVM3_1.fit(train[X3],train[Y3])
rbfSVM4_1=SVC(probability=True,C=10000,gamma=0.0001,kernel="rbf",class_weight='balanced')
rbfSVM4_1=rbfSVM4_1.fit(train[X4],train[Y4])
rbfSVM5_1=SVC(probability=True,C=100,gamma=0.01,kernel="rbf",class_weight='balanced')
rbfSVM5_1=rbfSVM5_1.fit(train[X5],train[Y5])
test_y_pre1_2=rbfSVM1_1.predict(test[X1])
print(metrics.accuracy_score(test[Y1],test_y_pre1_2))
print(metrics.classification_report(test[Y1],test_y_pre1_2))
test_y_pre2_2=rbfSVM2_1.predict(test[X2])
print(metrics.accuracy_score(test[Y2],test_y_pre2_2))
print(metrics.classification_report(test[Y2],test_y_pre2_2))
test_y_pre3_2=rbfSVM3_1.predict(test[X3])
print(metrics.accuracy_score(test[Y3],test_y_pre3_2))
print(metrics.classification_report(test[Y3],test_y_pre3_2))
test_y_pre4_2=rbfSVM4_1.predict(test[X4])
print(metrics.accuracy_score(test[Y4],test_y_pre4_2))
print(metrics.classification_report(test[Y4],test_y_pre4_2))
test_y_pre5_2=rbfSVM5_1.predict(test[X5])
print(metrics.accuracy_score(test[Y5],test_y_pre5_2))
print(metrics.classification_report(test[Y5],test_y_pre5_2))
#PLOT ROC CURVE
test_y_pre1_2=rbfSVM1_1.predict(test[X1])
testy__proba1_2=rbfSVM1_1.predict_proba(test[X1])[:,1]
fpr_rbfSVM1_1,tpr_rbfSVM1_1,thr =metrics.roc_curve(test[Y1],testy__proba1_2, pos_label=1)
auc=metrics.auc(fpr_rbfSVM1_1,tpr_rbfSVM1_1)
plt.figure(figsize=(6,6))
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr_rbfSVM1_1,tpr_rbfSVM1_1,"r",linewidth=1.5)
plt.xlabel("False Positive Rate",fontsize=15)
plt.ylabel("True Positive Rate",fontsize=15)
plt.xlim(0,1)
plt.ylim(0,1)
plt.title("rbfSVM1_1 ROC curve",fontsize=15)
plt.text(0.2,0.8,"auc="+str(round(auc,4)),fontsize=15)
plt.show()
test_y_pre2_2=rbfSVM2_1.predict(test[X2])
testy__proba2_2=rbfSVM2_1.predict_proba(test[X2])[:,1]
fpr_rbfSVM2_1,tpr_rbfSVM2_1,thr =metrics.roc_curve(test[Y2],testy__proba2_2, pos_label=1)
auc=metrics.auc(fpr_rbfSVM2_1,tpr_rbfSVM2_1)
plt.figure(figsize=(6,6))
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr_rbfSVM2_1,tpr_rbfSVM2_1,"r",linewidth=1.5)
plt.xlabel("False Positive Rate",fontsize=15)
plt.ylabel("True Positive Rate",fontsize=15)
plt.xlim(0,1)
plt.ylim(0,1)
plt.title("rbfSVM2_1 ROC curve",fontsize=15)
plt.text(0.2,0.8,"auc="+str(round(auc,4)),fontsize=15)
plt.show()
test_y_pre3_2=rbfSVM3_1.predict(test[X3])
testy__proba3_2=rbfSVM3_1.predict_proba(test[X3])[:,1]
fpr_rbfSVM3_1,tpr_rbfSVM3_1,thr =metrics.roc_curve(test[Y3],testy__proba3_2, pos_label=1)
auc=metrics.auc(fpr_rbfSVM3_1,tpr_rbfSVM3_1)
plt.figure(figsize=(6,6))
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr_rbfSVM3_1,tpr_rbfSVM3_1,"r",linewidth=1.5)
plt.xlabel("False Positive Rate",fontsize=15)
plt.ylabel("True Positive Rate",fontsize=15)
plt.xlim(0,1)
plt.ylim(0,1)
plt.title("rbfSVM3_1 ROC curve",fontsize=15)
plt.text(0.2,0.8,"auc="+str(round(auc,4)),fontsize=15)
plt.show()
test_y_pre4_2=rbfSVM4_1.predict(test[X4])
testy__proba4_2=rbfSVM4_1.predict_proba(test[X4])[:,1]
fpr_rbfSVM4_1,tpr_rbfSVM4_1,thr =metrics.roc_curve(test[Y4],testy__proba4_2, pos_label=1)
auc=metrics.auc(fpr_rbfSVM4_1,tpr_rbfSVM4_1)
plt.figure(figsize=(6,6))
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr_rbfSVM4_1,tpr_rbfSVM4_1,"r",linewidth=1.5)
plt.xlabel("False Positive Rate",fontsize=15)
plt.ylabel("True Positive Rate",fontsize=15)
plt.xlim(0,1)
plt.ylim(0,1)
plt.title("rbfSVM4_1 ROC curve",fontsize=15)
plt.text(0.2,0.8,"auc="+str(round(auc,4)),fontsize=15)
plt.show()
test_y_pre5_2=rbfSVM5_1.predict(test[X5])
testy__proba5_2=rbfSVM5_1.predict_proba(test[X5])[:,1]
fpr_rbfSVM5_1,tpr_rbfSVM5_1,thr =metrics.roc_curve(test[Y5],testy__proba5_2, pos_label=1)
auc=metrics.auc(fpr_rbfSVM5_1,tpr_rbfSVM5_1)
plt.figure(figsize=(6,6))
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr_rbfSVM5_1,tpr_rbfSVM5_1,"r",linewidth=1.5)
plt.xlabel("False Positive Rate",fontsize=15)
plt.ylabel("True Positive Rate",fontsize=15)
plt.xlim(0,1)
plt.ylim(0,1)
plt.title("rbfSVM5_1 ROC curve",fontsize=15)
plt.text(0.2,0.8,"auc="+str(round(auc,4)),fontsize=15)
plt.show()
#plot_learning_curve
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import learning_curve
def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None,n_jobs=1, train_sizes=np.linspace(.1, 1.0, 5)):
    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    train_sizes, train_scores, test_scores = learning_curve(estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()
    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,train_scores_mean + train_scores_std, alpha=0.1,color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",label="Cross-validation score")
    plt.legend(loc="best")
    return plt
 
title = r"Learning Curves (SVM1)"
cv = ShuffleSplit(n_splits=10, test_size=0.2, random_state=0)
estimator = rbfSVM1_1 
plot_learning_curve(estimator, title, train[X1],train[Y1], ylim=(0.7, 1.01), cv=cv, n_jobs=1)
plt.show()
```
