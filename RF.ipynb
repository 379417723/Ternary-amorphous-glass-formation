```python
#######RF
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei']
import scipy as sp
%matplotlib inline
%config inlinebackend.figure_format="retina"
from matplotlib.font_manager import FontProperties
from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm
import matplotlib as mpl
import seaborn as sns
from sklearn import tree,metrics
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
from sklearn.datasets import make_classification
import graphviz
import pydotplus
from io import StringIO
from IPython.display import Image

data=pd.read_csv("*:/****/Ternary_Glassformation_dataset.csv")
train,test = train_test_split(data, test_size=0.2,random_state=2)
print(data.shape)
print(train.shape)
print(test.shape)
data_pro=[train,test]
print("data.info：\n",data.info())
print(data.columns.values)
#1G-90
Y1=['GFA']
X1=['a1','a2','a3','Ri1','Ri2','Ri3','E1','E2','E3','Tb1','Tb2','Tb3','Tm1','Tm2','Tm3','ρ1','ρ2','ρ3','Aw1','Aw2','Aw3', 'Rc1','Rc2','Rc3','Ra1','Ra2','Ra3','IE1','IE2','IE3','Eea1','Eea2','Eea3','Ia1','Ia2','Ib1','Ib2','Ic1','Ic2','α1','α2','α3','Cp1','Cp2','Cp3','λ1','λ2','λ3','σ1','σ2','σ3','Hf1','Hf2','Hf3','Hv1','Hv2','Hv3',
'χ1','χ2','χ3','AN1','AN2','AN3','P1','P2','P3','GP1','GP2','GP3','VEC1','VEC2','VEC3','sVEC1','sVEC2','sVEC3','pVEC1','pVEC2','pVEC3','dVEC1','dVEC2','dVEC3','W1','W2','W3','LP1','LP2','LP3','∆H12','∆H13','∆H23']
print(train[Y1].shape)
print(train[X1].shape)
#2G-32
Y2=['GFA']
X2=['Ri','E','Tb','Tm','ρ','Aw', 'Rc','Ra','IE','Eea','I1','I2','α','Cp','λ','σ','Hf','Hv','χ','AN','P','GP','VEC','sVEC','pVEC','dVEC','W','LP','△Hmix','△Sc','δ','VA']
print(train[Y2].shape)
print(train[X2].shape)

#FEATURE SELECTION
#PEARSON'S COEFFICIENT
import matplotlib
matplotlib.rcParams['axes.unicode_minus']=False
corr_features=train[X2].corr(method='pearson')
#sns.heatmap(corr,cmap='RdBu', linewidths = 0.01, ax = 0)
plt.figure(figsize=(18,8))
sns.heatmap(corr_features,annot=True,cmap='RdBu_r')
#Cv
mean=numpy.mean(data[X2])
std=numpy.std(data[X2])
var=numpy.var(data[X2])
cv=std/mean
print(cv)
print(var)
#3G-18
Y3=['GFA']#PEARSON<0.7
X3=['Ri','E','Tb','Aw','Rc','IE','Eea','α','σ','Hf','GP','sVEC','pVEC','dVEC','LP','△Hmix','△Sc','δ']
print(train[Y3].shape)
print(train[X3].shape)
#4G-18
Y4=['GFA']#Cv>0.3
X4=['E','ρ','Aw','Eea','α','Cp','λ','σ','Hf','AN','VEC','pVEC','dVEC','LP','△Hmix','△Sc','δ','VA']
print(train[Y4].shape)
print(train[X4].shape)
#5G-12
Y5=['GFA']#PEARSON<0.7 and Cv>0.3
X5=['Aw','Eea','α','σ','Hf','pVEC','dVEC','LP','△Hmix','△Sc','δ','VA',]
print(train[Y5].shape)
print(train[X5].shape)
traindata_x1,valdata_x1,traindata_y1,valdata_y1=train_test_split(train[X1],train[Y1],test_size=0.25,random_state=1)
print(traindata_x1.shape)
print(valdata_x1.shape)
traindata_x2,valdata_x2,traindata_y2,valdata_y2=train_test_split(train[X2],train[Y2],test_size=0.25,random_state=1)
print(traindata_x2.shape)
print(valdata_x2.shape)
traindata_x3,valdata_x3,traindata_y3,valdata_y3=train_test_split(train[X3],train[Y3],test_size=0.25,random_state=1)
print(traindata_x3.shape)
print(valdata_x3.shape)
traindata_x4,valdata_x4,traindata_y4,valdata_y4=train_test_split(train[X4],train[Y4],test_size=0.25,random_state=1)
print(traindata_x4.shape)
print(valdata_x4.shape)
traindata_x5,valdata_x5,traindata_y5,valdata_y5=train_test_split(train[X5],train[Y5],test_size=0.25,random_state=1)
print(traindata_x5.shape)
print(valdata_x5.shape)
#RF
RF1=RandomForestClassifier(random_state=1)
RF1.fit(traindata_x1,traindata_y1)
RF2=RandomForestClassifier(random_state=1)
RF2.fit(traindata_x2,traindata_y2)
RF3=RandomForestClassifier(random_state=1)
RF3.fit(traindata_x3,traindata_y3)
RF4=RandomForestClassifier(random_state=1)
RF4.fit(traindata_x4,traindata_y4)
RF5=RandomForestClassifier(random_state=1)
RF5.fit(traindata_x5,traindata_y5)
valdatapre_y1_1=RF1.predict(valdata_x1)
print(metrics.accuracy_score(valdata_y1,valdatapre_y1_1))
print(metrics.classification_report(valdata_y1,valdatapre_y1_1))
valdatapre_y2_1=RF2.predict(valdata_x2)
print(metrics.accuracy_score(valdata_y2,valdatapre_y2_1))
print(metrics.classification_report(valdata_y2,valdatapre_y2_1))
valdatapre_y3_1=RF3.predict(valdata_x3)
print(metrics.accuracy_score(valdata_y3,valdatapre_y3_1))
print(metrics.classification_report(valdata_y3,valdatapre_y3_1))
valdatapre_y4_1=RF4.predict(valdata_x4)
print(metrics.accuracy_score(valdata_y4,valdatapre_y4_1))
print(metrics.classification_report(valdata_y4,valdatapre_y4_1))
valdatapre_y5_1=RF5.predict(valdata_x5)
print(metrics.accuracy_score(valdata_y5,valdatapre_y5_1))
print(metrics.classification_report(valdata_y5,valdatapre_y5_1))
#PLOT ROC CURVE
valdatapre_y1_1=RF1.predict(valdata_x1)
valdatay__proba1=RF1.predict_proba(valdata_x1)[:,1]
fpr_RF1,tpr_RF1,thr =metrics.roc_curve(valdata_y1,valdatay__proba1, pos_label=1)
auc=metrics.auc(fpr_RF1,tpr_RF1)
plt.figure(figsize=(6,6))
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr_RF1,tpr_RF1,"r",linewidth=1.5)
plt.xlabel("False Positive Rate",fontsize=15)
plt.ylabel("True Positive Rate",fontsize=15)
plt.xlim(0,1)
plt.ylim(0,1)
plt.title("RF1 ROC curve",fontsize=15)
plt.text(0.2,0.8,"auc="+str(round(auc,4)),fontsize=15)
plt.show()
valdatapre_y2_1=RF2.predict(valdata_x2)
valdatay__proba2=RF2.predict_proba(valdata_x2)[:,1]
fpr_RF2,tpr_RF2,thr =metrics.roc_curve(valdata_y2,valdatay__proba2, pos_label=1)
auc=metrics.auc(fpr_RF2,tpr_RF2)
plt.figure(figsize=(6,6))
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr_RF2,tpr_RF2,"r",linewidth=1.5)
plt.xlabel("False Positive Rate",fontsize=15)
plt.ylabel("True Positive Rate",fontsize=15)
plt.xlim(0,1)
plt.ylim(0,1)
plt.title("RF2 ROC curve",fontsize=15)
plt.text(0.2,0.8,"auc="+str(round(auc,4)),fontsize=15)
plt.show()
valdatapre_y3_1=RF3.predict(valdata_x3)
valdatay__proba3=RF3.predict_proba(valdata_x3)[:,1]
fpr_RF3,tpr_RF3,thr =metrics.roc_curve(valdata_y3,valdatay__proba3, pos_label=1)
auc=metrics.auc(fpr_RF3,tpr_RF3)
plt.figure(figsize=(6,6))
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr_RF3,tpr_RF3,"r",linewidth=1.5)
plt.xlabel("False Positive Rate",fontsize=15)
plt.ylabel("True Positive Rate",fontsize=15)
plt.xlim(0,1)
plt.ylim(0,1)
plt.title("RF3 ROC curve",fontsize=15)
plt.text(0.2,0.8,"auc="+str(round(auc,4)),fontsize=15)
plt.show()
valdatapre_y4_1=RF4.predict(valdata_x4)
valdatay__proba4=RF4.predict_proba(valdata_x4)[:,1]
fpr_RF4,tpr_RF4,thr =metrics.roc_curve(valdata_y4,valdatay__proba4, pos_label=1)
auc=metrics.auc(fpr_RF4,tpr_RF4)
plt.figure(figsize=(6,6))
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr_RF4,tpr_RF4,"r",linewidth=1.5)
plt.xlabel("False Positive Rate",fontsize=15)
plt.ylabel("True Positive Rate",fontsize=15)
plt.xlim(0,1)
plt.ylim(0,1)
plt.title("RF4 ROC curve",fontsize=15)
plt.text(0.2,0.8,"auc="+str(round(auc,4)),fontsize=15)
plt.show()
valdatapre_y5_1=RF5.predict(valdata_x5)
valdatay__proba5=RF5.predict_proba(valdata_x5)[:,1]
fpr_RF5,tpr_RF5,thr =metrics.roc_curve(valdata_y5,valdatay__proba5, pos_label=1)
auc=metrics.auc(fpr_RF5,tpr_RF5)
plt.figure(figsize=(6,6))
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr_RF5,tpr_RF5,"r",linewidth=1.5)
plt.xlabel("False Positive Rate",fontsize=15)
plt.ylabel("True Positive Rate",fontsize=15)
plt.xlim(0,1)
plt.ylim(0,1)
plt.title("RF5 ROC curve",fontsize=15)
plt.text(0.2,0.8,"auc="+str(round(auc,4)),fontsize=15)
plt.show()

#GridSearchCV
from sklearn.model_selection import GridSearchCV
rfgs=RandomForestClassifier(oob_score=True,random_state=1)
n_estimators=[50,100,200,500,700,1000,2000]
max_depth=[3,5,8,10,15,18,20,25,28,30,40]
para_grid=[{"n_estimators":n_estimators,"max_depth":max_depth}]
RF1_GridSearch=GridSearchCV(estimator=rfgs,param_grid=para_grid,cv=5,n_jobs=4)
RF2_GridSearch=GridSearchCV(estimator=rfgs,param_grid=para_grid,cv=5,n_jobs=4)
RF3_GridSearch=GridSearchCV(estimator=rfgs,param_grid=para_grid,cv=5,n_jobs=4)
RF4_GridSearch=GridSearchCV(estimator=rfgs,param_grid=para_grid,cv=5,n_jobs=4)
RF5_GridSearch=GridSearchCV(estimator=rfgs,param_grid=para_grid,cv=5,n_jobs=4)
print(RF1_GridSearch.fit(traindata_x1,traindata_y1))
print(RF2_GridSearch.fit(traindata_x2,traindata_y2))
print(RF3_GridSearch.fit(traindata_x3,traindata_y3))
print(RF4_GridSearch.fit(traindata_x4,traindata_y4))
print(RF5_GridSearch.fit(traindata_x5,traindata_y5))
print("best_params：",RF1_GridSearch.best_params_)
print("best_score：",RF1_GridSearch.best_score_)
print("best_estimator：",RF1_GridSearch.best_estimator_)
print("best_params：",RF2_GridSearch.best_params_)
print("best_score：",RF2_GridSearch.best_score_)
print("best_estimator：",RF2_GridSearch.best_estimator_)
print("best_params：",RF3_GridSearch.best_params_)
print("best_score：",RF3_GridSearch.best_score_)
print("best_estimator：",RF3_GridSearch.best_estimator_)
print("best_params：",RF4_GridSearch.best_params_)
print("best_score：",RF4_GridSearch.best_score_)
print("best_estimator：",RF4_GridSearch.best_estimator_)
print("best_params：",RF5_GridSearch.best_params_)
print("best_score：",RF5_GridSearch.best_score_)
print("best_estimator：",RF5_GridSearch.best_estimator_)
#RETRAINING WITH TUNED PARAMETERS
RF1_1=RandomForestClassifier(n_estimators=50,max_depth=20,oob_score=True,random_state=1,class_weight=({0:750, 1:600}))
RF1_1.fit(train[X1],train[Y1])
print("oob score:",round(RF1_1.oob_score_,4)*100,"%")
print(RF1_1.feature_importances_)
RF2_1=RandomForestClassifier(n_estimators=500,max_depth=25,oob_score=True,random_state=1,class_weight=({0:750, 1:600}))
RF2_1.fit(train[X2],train[Y2])
print("oob score:",round(RF2_1.oob_score_,4)*100,"%")
print(RF2_1.feature_importances_)
RF3_1=RandomForestClassifier(n_estimators=200,max_depth=18,oob_score=True,random_state=1,class_weight=({0:750, 1:600}))
RF3_1.fit(train[X3],train[Y3])
print("oob score:",round(RF3_1.oob_score_,4)*100,"%")
print(RF3_1.feature_importances_
RF4_1=RandomForestClassifier(n_estimators=700,max_depth=25,oob_score=True,random_state=1,class_weight=({0:750, 1:600}))
RF4_1.fit(train[X4],train[Y4])
print("oob score:",round(RF4_1.oob_score_,4)*100,"%")
print(RF4_1.feature_importances_)
RF5_1=RandomForestClassifier(n_estimators=100,max_depth=15,oob_score=True,random_state=1,class_weight=({0:750, 1:600}))
RF5_1.fit(train[X5],train[Y5])
print("oob score:",round(RF4_1.oob_score_,4)*100,"%")
print(RF5_1.feature_importances_)
test_y_pre1_2=RF1_1.predict(test[X1])
print(metrics.accuracy_score(test[Y1],test_y_pre1_2))
print(metrics.classification_report(test[Y1],test_y_pre1_2))
test_y_pre2_2=RF2_1.predict(test[X2])
print(metrics.accuracy_score(test[Y2],test_y_pre2_2))
print(metrics.classification_report(test[Y2],test_y_pre2_2))
test_y_pre3_2=RF3_1.predict(test[X3])
print(metrics.accuracy_score(test[Y3],test_y_pre3_2))
print(metrics.classification_report(test[Y3],test_y_pre3_2))
test_y_pre4_2=RF4_1.predict(test[X4])
print(metrics.accuracy_score(test[Y4],test_y_pre4_2))
print(metrics.classification_report(test[Y4],test_y_pre4_2))
test_y_pre5_2=RF5_1.predict(test[X5])
print(metrics.accuracy_score(test[Y5],test_y_pre5_2))
print(metrics.classification_report(test[Y5],test_y_pre5_2))
#PLOT ROC CURVE
test_y_pre1_2=RF1_1.predict(test[X1])
testy__proba1_2=RF1_1.predict_proba(test[X1])[:,1]
fpr_RF1_1,tpr_RF1_1,thr =metrics.roc_curve(test[Y1],testy__proba1_2, pos_label=1)
auc=metrics.auc(fpr_RF1_1,tpr_RF1_1)
plt.figure(figsize=(6,6))
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr_RF1_1,tpr_RF1_1,"r",linewidth=1.5)
plt.xlabel("False Positive Rate",fontsize=15)
plt.ylabel("True Positive Rate",fontsize=15)
plt.xlim(0,1)
plt.ylim(0,1)
plt.title("RF1_1 ROC curve",fontsize=15)
plt.text(0.2,0.8,"auc="+str(round(auc,4)),fontsize=15)
plt.show()
test_y_pre2_2=RF2_1.predict(test[X2])
testy__proba2_2=RF2_1.predict_proba(test[X2])[:,1]
fpr_RF2_1,tpr_RF2_1,thr =metrics.roc_curve(test[Y2],testy__proba2_2, pos_label=1)
auc=metrics.auc(fpr_RF2_1,tpr_RF2_1)
plt.figure(figsize=(6,6))
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr_RF2_1,tpr_RF2_1,"r",linewidth=1.5)
plt.xlabel("False Positive Rate",fontsize=15)
plt.ylabel("True Positive Rate",fontsize=15)
plt.xlim(0,1)
plt.ylim(0,1)
plt.title("RF2_1 ROC curve",fontsize=15)
plt.text(0.2,0.8,"auc="+str(round(auc,4)),fontsize=15)
plt.show()
test_y_pre3_2=RF3_1.predict(test[X3])
testy__proba3_2=RF3_1.predict_proba(test[X3])[:,1]
fpr_RF3_1,tpr_RF3_1,thr =metrics.roc_curve(test[Y3],testy__proba3_2, pos_label=1)
auc=metrics.auc(fpr_RF3_1,tpr_RF3_1)
plt.figure(figsize=(6,6))
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr_RF3_1,tpr_RF3_1,"r",linewidth=1.5)
plt.xlabel("False Positive Rate",fontsize=15)
plt.ylabel("True Positive Rate",fontsize=15)
plt.xlim(0,1)
plt.ylim(0,1)
plt.title("RF3_1 ROC curve",fontsize=15)
plt.text(0.2,0.8,"auc="+str(round(auc,4)),fontsize=15)
plt.show()
test_y_pre4_2=RF4_1.predict(test[X4])
testy__proba4_2=RF4_1.predict_proba(test[X4])[:,1]
fpr_RF4_1,tpr_RF4_1,thr =metrics.roc_curve(test[Y4],testy__proba4_2, pos_label=1)
auc=metrics.auc(fpr_RF4_1,tpr_RF4_1)
plt.figure(figsize=(6,6))
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr_RF4_1,tpr_RF4_1,"r",linewidth=1.5)
plt.xlabel("False Positive Rate",fontsize=15)
plt.ylabel("True Positive Rate",fontsize=15)
plt.xlim(0,1)
plt.ylim(0,1)
plt.title("RF4_1 ROC curve",fontsize=15)
plt.text(0.2,0.8,"auc="+str(round(auc,4)),fontsize=15)
plt.show()
test_y_pre5_2=RF5_1.predict(test[X5])
testy__proba5_2=RF5_1.predict_proba(test[X5])[:,1]
fpr_RF5_1,tpr_RF5_1,thr =metrics.roc_curve(test[Y5],testy__proba5_2, pos_label=1)
auc=metrics.auc(fpr_RF5_1,tpr_RF5_1)
plt.figure(figsize=(6,6))
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr_RF5_1,tpr_RF5_1,"r",linewidth=1.5)
plt.xlabel("False Positive Rate",fontsize=15)
plt.ylabel("True Positive Rate",fontsize=15)
plt.xlim(0,1)
plt.ylim(0,1)
plt.title("RF5_1 ROC curve",fontsize=15)
plt.text(0.2,0.8,"auc="+str(round(auc,4)),fontsize=15)
plt.show()
#PLOT LEARNING CURVE
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import learning_curve
def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None,n_jobs=1, train_sizes=np.linspace(.1, 1.0, 10)):
    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training size")
    plt.ylabel("Accuracy")
    train_sizes, train_scores, test_scores = learning_curve(estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()
    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,train_scores_mean + train_scores_std, alpha=0.1,color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",label="Training")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",label="Test")
    plt.legend(loc="best") 
    numpy.savetxt('train_scores_mean.csv',train_scores_mean, delimiter = ',') 
    numpy.savetxt('test_scores_mean.csv',test_scores_mean, delimiter = ',') 
    return plt
title = r"Learning Curves (RF5_1)"
cv = ShuffleSplit(n_splits=10, test_size=0.2, random_state=0)
estimator =RF5_1
plot_learning_curve(estimator, title, data[X5],data[Y5], ylim=(0.7, 1.01), cv=cv, n_jobs=1)
plt.show()
#CONFUSION_MATRIX
confm=metrics.confusion_matrix(test[Y1],test_y_pre1_2)
sns.heatmap(confm.T,square=True,annot=True,fmt='d',cbar=False)
plt.xlabel('True Label')
plt.ylabel('Predict Label')
plt.title('x1')
confm=metrics.confusion_matrix(test[Y2],test_y_pre2_2)
sns.heatmap(confm.T,square=True,annot=True,fmt='d',cbar=False)
plt.xlabel('True Label')
plt.ylabel('Predict Label')
plt.title('x2')
confm=metrics.confusion_matrix(test[Y3],test_y_pre3_2)
sns.heatmap(confm.T,square=True,annot=True,fmt='d',cbar=False)
plt.xlabel('True Label',fontproperties = 'Times New Roman', size = 14)
plt.ylabel('Predict Label',fontproperties = 'Times New Roman', size = 14)
plt.title('X3',fontproperties = 'Times New Roman', size = 14)
confm=metrics.confusion_matrix(test[Y4],test_y_pre4_2)
sns.heatmap(confm.T,square=True,annot=True,fmt='d',cbar=False)
plt.xlabel('True Label')
plt.ylabel('Predict Label')
plt.title('x4')
confm=metrics.confusion_matrix(test[Y5],test_y_pre5_2)
sns.heatmap(confm.T,square=True,annot=True,fmt='d',cbar=False)
plt.xlabel('True Label')
plt.ylabel('Predict Label')
plt.title('x5')
      
#COMPOSITION SUBSPACES
No_Cu_Ni_Ti=pd.read_csv("*:/****/no_Cu_Ni_Ti.csv")#dataset without Cu_Ni_Ti
Cu_Ni_Ti=pd.read_csv("*:/****/Cu_Ni_Ti.csv")#dataset of Cu_Ni_Ti
print(No_Cu_Ni_Ti.shape)
print(Cu_Ni_Ti.shape)
RF_data=RandomForestClassifier(n_estimators=100,max_depth=15,oob_score=True,random_state=1)
RF_No_Cu_Ni_Ti=RF_data.fit(No_Cu_Ni_Ti[X5],No_Cu_Ni_Ti[Y5])
Cu_Ni_Ti_prey=RF_No_Cu_Ni_Ti.predict(Cu_Ni_Ti[X5])  
numpy.savetxt('Cu_Ni_Tiprey.csv',Cu_Ni_Ti_prey, delimiter = ',') 
Cu_Ni_Ti_probay=RF_No_Cu_Ni_Ti.predict_proba(Cu_Ni_Ti[X5])[:,1]
numpy.savetxt('Cu_Ni_Ti_probay.csv',Cu_Ni_Ti_probay, delimiter = ',') 
print(Cu_Ni_Ti_probay.shape)
no_Mg_Nd_Cu=pd.read_csv("*:/***/no_Mg_Nd_Cu.csv")#dataset without Mg_Nd_Cu
Mg_Nd_Cu=pd.read_csv("*:/***/Mg_Nd_Cu.csv")#dataset of Mg_Nd_Cu
print(no_Mg_Nd_Cu.shape)
print(Mg_Nd_Cu.shape)
RF_data=RandomForestClassifier(n_estimators=100,max_depth=15,oob_score=True,random_state=1)
RF_no_Mg_Nd_Cu=RF_data.fit(no_Mg_Nd_Cu[X5],no_Mg_Nd_Cu[Y5])
Mg_Nd_Cu_prey=RF_no_Mg_Nd_Cu.predict(Mg_Nd_Cu[X5])
numpy.savetxt('Mg_Nd_Cu_prey.csv',Mg_Nd_Cu_prey, delimiter = ',') 
Mg_Nd_Cu_probay=RF_no_Mg_Nd_Cu.predict_proba(Mg_Nd_Cu[X5])[:,1]
numpy.savetxt('Mg_Nd_Ni_probay.csv',Mg_Nd_Cu_probay, delimiter = ',') 
print(Mg_Nd_Cu_probay.shape)
no_Ni_Si_Ti=pd.read_csv("*:/***/no_Ni_Si_Ti.csv")
Ni_Si_Ti=pd.read_csv("*:/***/Ni_Si_Ti.csv")
print(no_Ni_Si_Ti.shape)
print(Ni_Si_Ti.shape)
RF_data=RandomForestClassifier(n_estimators=100,max_depth=15,oob_score=True,random_state=1)
RF_no_Ni_Si_Ti=RF_data.fit(no_Ni_Si_Ti[X5],no_Ni_Si_Ti[Y5])
Ni_Si_Ti_prey=RF_no_Ni_Si_Ti.predict(Ni_Si_Ti[X5])
numpy.savetxt('Ni_Si_Ti_prey.csv',Ni_Si_Ti_prey, delimiter = ',') 
Ni_Si_Ti_probay=RF_no_Ni_Si_Ti.predict_proba(Ni_Si_Ti[X5])[:,1]
numpy.savetxt('Ni_Si_Ti_probay.csv',Ni_Si_Ti_probay, delimiter = ',') 
print(Ni_Si_Ti_probay.shape)
      
#multicomponent alloy test
RFdata=RandomForestClassifier(n_estimators=100,max_depth=15,oob_score=True,random_state=1)
RFdata.fit(data[X5],data[Y5])
multi_alloy=pd.read_csv("*:/****/multi_alloy.csv")
print(multi_alloy.shape)
multi_alloy_prey=RFdata.predict(multi_alloy[X5])
print(metrics.accuracy_score(multi_alloy[Y5],multi_alloy_prey))
print(metrics.classification_report(multi_alloy[Y5],multi_alloy_prey))
#CONFUSION_MATRIX
confm=metrics.confusion_matrix(multi_alloy[Y5],multi_alloy_prey)
sns.heatmap(confm.T,square=True,annot=True,fmt='d',cbar=False)
plt.xlabel('True Label',fontproperties = 'Times New Roman', size = 14)
plt.ylabel('Predict Label',fontproperties = 'Times New Roman', size = 14)
plt.title('multi_alloy',fontproperties = 'Times New Roman', size = 14)
```
